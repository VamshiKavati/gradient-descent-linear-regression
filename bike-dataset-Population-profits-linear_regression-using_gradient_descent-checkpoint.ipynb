```python
# Loading the libraries

import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
import seaborn as sns
import pandas as pd
```

```python
#importing data

data = pd.read_csv('bike-dataset.txt')
print(data.head())

x = data['Population'].values
y = data['Profit']
```

```python
# finding global minima

def gradient_descent(x, y, m_curr, b_curr):
    iterations = 10000
    lr = 0.01
    n = len(x)


    for i in range(iterations):
        y_pred = m_curr * x +b_curr
        cost = (1/n) *sum((y-y_pred)**2)
        md = -(2/n) *sum(x*(y-y_pred))
        cd = -(2/n) *sum(y-y_pred)
        m_curr = m_curr - lr * md
        b_curr = b_curr - lr * cd
        print("m {0} , b {1}, iteration {2}, cost {3}".format(m_curr, b_curr, i, cost))
        
    return m_curr, b_curr

m_curr = b_curr = 0
m_curr, b_curr = gradient_descent(x,y, m_curr, b_curr)
```

```python
print(m_curr, b_curr)
```

```python
# Plot the best fit line and scatter pints
X = np.linspace((min(x) - 1),(max(x)+1),1000)
y_pred = m_curr * X +b_curr

plt.scatter(x,y, color = "r", label="Scatter Points")
plt.plot(X,y_pred, color = "g", label="Regression Points")

plt.xlabel = ("Populations")
plt.ylabel = ("Profits")
plt.legend()
plt.show()
```
